## Excercise 1

Plot the $pdf(d)$, $cdf(p)$ and a histogram of a randomly generated sample form a gamma, $GA(\mu, \sigma)$, distribution with $\mu = 3$ and $\sigme = 0.5$, using the gamlss.dist functions 
* dGA()
* pGA()
* qGA() and 
* rGA()

respectively.

```r
mu <- 3
sigma <- .5
curve (dGA(y, mu, sigma), 0.01, 10, xname="y", ylab="f(y)", main = "(a)") # pdf
curve (pGA(y, mu, sigma), 0.01, 10, xname="y", ylab="f(y)", main = "(b)") # cdf
set.seed (125)
y<- rGA(1000, mu, sigma) # random sample hist(y, col = "lightgray", main= "(d)")
```
